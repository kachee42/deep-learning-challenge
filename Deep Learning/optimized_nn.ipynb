```python
# Import our dependencies
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import pandas as pd
import tensorflow as tf

#  Import and read the charity_data.csv.
import pandas as pd 
application_df = pd.read_csv("https://static.bc-edx.com/data/dl-1-2/m21/lms/starter/charity_data.csv")
application_df.head()
```

```python
# Drop the non-beneficial ID columns, 'EIN' and 'NAME'.
application_df = application_df.drop(["EIN", "NAME"], axis=1)
application_df.head()
```

```python
# Determine the number of unique values in each column.
application_df.nunique()
```

```python
# Look at APPLICATION_TYPE value counts for binning
application_df["APPLICATION_TYPE"].value_counts()
```

```python
application_df.info()
```

```python
# Choose a cutoff value and create a list of application types to be replaced
# use the variable name `application_types_to_replace`
application_types_10_100 = ["T13", "T12", "T2"]
application_types_lt_10 = ["T25", "T14", "T29", "T15", "T17"]

# Replace in dataframe
for app in application_types_10_100:
    application_df['APPLICATION_TYPE'] = application_df['APPLICATION_TYPE'].replace(app,"APP_TYPE_appears_10_100")

for app in application_types_lt_10:
    application_df['APPLICATION_TYPE'] = application_df['APPLICATION_TYPE'].replace(app,"APP_TYPE_appears_lt_10")

# Check to make sure binning was successful
application_df['APPLICATION_TYPE'].value_counts()
```

```python
# Look at CLASSIFICATION value counts for binning
application_df["CLASSIFICATION"].value_counts()
```

```python
# You may find it helpful to look at CLASSIFICATION value counts >1
application_df['CLASSIFICATION'].value_counts()[application_df['CLASSIFICATION'].value_counts()> 1]
```

```python
# Choose a cutoff value and create a list of classifications to be replaced
# use the variable name `classifications_to_replace`
classifications_1 = ["C2190", "C4200", "C2600", "C5200", 
                     "C1370", "C1248", "C6100", "C1820", 
                     "C1900", "C1236", "C3700", "C2570", 
                     "C1580", "C1245", "C2500", "C1570",
                     "C1283", "C2380", "C1732", "C1728",
                     "C2170", "C4120", "C8210", "C2561",
                     "C4500", "C2150"]

classifications_2_10 = ["C1238", "C1278", "C1235", "C1237",
                        "C7210", "C2400", "C1720", "C4100",
                        "C1257", "C1600", "C1260", "C2710",
                        "C0", "C3200", "C1234", "C1246",
                        "C1267", "C1256"]

classifications_11_20 = ["C8000", "C7120", "C1500", "C1800",
                         "C6000", "C1250", "C8200"]

classifications_30_50 = ["C1280", "C1230", "C1400", "C7200",
                         "C2300", "C1240"]

# Replace in dataframe
for cls in classifications_1:
    application_df['CLASSIFICATION'] = application_df['CLASSIFICATION'].replace(cls,"CLASS_ultra_rare")

for cls in classifications_2_10:
    application_df['CLASSIFICATION'] = application_df['CLASSIFICATION'].replace(cls,"CLASS_very_rare")

for cls in classifications_11_20:
    application_df['CLASSIFICATION'] = application_df['CLASSIFICATION'].replace(cls,"CLASS_rare")

for cls in classifications_30_50:
    application_df['CLASSIFICATION'] = application_df['CLASSIFICATION'].replace(cls,"CLASS_uncommon")
    
# Check to make sure binning was successful
application_df['CLASSIFICATION'].value_counts()
```

```python
# Convert categorical data to numeric with `pd.get_dummies`
categorical_dummies = pd.get_dummies(application_df[["APPLICATION_TYPE", "AFFILIATION", "CLASSIFICATION", "USE_CASE", "ORGANIZATION", "INCOME_AMT", "SPECIAL_CONSIDERATIONS"]])
application_df = pd.concat([application_df, categorical_dummies], axis=1)
application_df = application_df.drop(columns=["APPLICATION_TYPE", "AFFILIATION", "CLASSIFICATION", "USE_CASE", "ORGANIZATION", "INCOME_AMT", "SPECIAL_CONSIDERATIONS"])
application_df.head()
```

```python
# Split our preprocessed data into our features and target arrays
y = application_df["IS_SUCCESSFUL"].values
X = application_df.drop(["IS_SUCCESSFUL"], 1).values

# Split the preprocessed data into a training and testing dataset
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=10)
```

```python
# Create a StandardScaler instances
scaler = StandardScaler()

# Fit the StandardScaler
X_scaler = scaler.fit(X_train)

# Scale the data
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

## Optimize keras model

```python
def create_model(hp):
    nn_model = tf.keras.models.Sequential()

    number_input_features = len(X_train[0])

    # Allow kerastuner to decide which activation function to use in hidden layers
    activation = hp.Choice('activation',['relu','tanh', 'sigmoid'])
    
    # Allow kerastuner to decide number of neurons in first layer
    nn_model.add(tf.keras.layers.Dense(units=hp.Int('first_units',
        min_value=1,
        max_value=100,
        step=1), activation=activation, input_dim=number_input_features))

    # Allow kerastuner to decide number of hidden layers and neurons in hidden layers
    for i in range(hp.Int('num_layers', 1, 6)):
        nn_model.add(tf.keras.layers.Dense(units=hp.Int('units_' + str(i),
            min_value=1,
            max_value=100,
            step=1),
            activation=activation))
    
    nn_model.add(tf.keras.layers.Dense(units=1, activation="sigmoid"))

    # Compile the model
    nn_model.compile(loss="binary_crossentropy", optimizer='adam', metrics=["accuracy"])
    
    return nn_model
```

```python
# Import the kerastuner library
import keras_tuner as kt

tuner = kt.Hyperband(
    create_model,
    objective="val_accuracy",
    max_epochs=50,
    hyperband_iterations=2)
```

```python
tuner.search(X_train_scaled, y_train, epochs=50, validation_data=(X_test_scaled, y_test))
```

## Compile, Train and Evaluate the Model

```python
# Define the model - deep neural net, i.e., the number of input features and hidden nodes for each layer.
number_input_features = len(X_train[0])
layer1              = 94
hidden_nodes_layer1 = 4
hidden_nodes_layer2 = 35
hidden_nodes_layer3 = 60
hidden_nodes_layer4 = 13
hidden_nodes_layer5 = 29
hidden_nodes_layer6 = 1

nn = tf.keras.models.Sequential()

# First layer
nn.add(tf.keras.layers.Dense(units=layer1, input_dim=number_input_features, activation="tanh"))

# First hidden layer
nn.add(tf.keras.layers.Dense(units=hidden_nodes_layer1, activation="tanh"))

# Second hidden layer
nn.add(tf.keras.layers.Dense(units=hidden_nodes_layer2, activation="tanh"))

# Third hidden layer
nn.add(tf.keras.layers.Dense(units=hidden_nodes_layer3, activation="tanh"))

# # Fourth hidden layer
nn.add(tf.keras.layers.Dense(units=hidden_nodes_layer4, activation="tanh"))

# # Fifth hidden layer
nn.add(tf.keras.layers.Dense(units=hidden_nodes_layer5, activation="tanh"))

# # Sixth hidden layer
nn.add(tf.keras.layers.Dense(units=hidden_nodes_layer6, activation="tanh"))

# Output layer
nn.add(tf.keras.layers.Dense(units=1, activation="sigmoid"))

# Check the structure of the model
nn.summary()
```

```python
# Compile the model
nn.compile(loss="binary_crossentropy", optimizer="adam", metrics=["accuracy"])
```

```python
# Train the model
fit_model = nn.fit(X_train_scaled, y_train, epochs=100)
```

```python
# Evaluate the model using the test data
model_loss, model_accuracy = nn.evaluate(X_test_scaled,y_test,verbose=2)
print(f"Loss: {model_loss}, Accuracy: {model_accuracy}")
```
